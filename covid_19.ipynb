```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = dataset.drop(["Lat", "Long"], axis=1, inplace = True)
```

```python
#aggregating the data
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
#Calculating the maximum infection rate for all the countries
countries = list(corona_dataset_aggregated.index)
#Dear readers, here I was repeadely getting a type error and value error, which wasted most of my time and also confused me a lot.
#But then after wasting one day I realised that here python particularly pandas was treating the data frame columns and string as same.
#Due to this it also read the "Country/Region" column with other numeric value columns, which was the causing the error.
# SO to avoid this I just used "select_dtypes" method to select only the numeric data. 
#Thank you for reading
numeric_data = corona_dataset_aggregated.select_dtypes(include = 'number')

corona_dataset_aggregated["Max Infection Rate"] = (numeric_data.diff(axis=1).max(axis=1))

print(type(corona_dataset_aggregated))
```

```python
corona_dataset_aggregated
```

```python
#Creating a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max Infection Rate"])
```

```python
corona_data
```

```python
#importing the second dataset
world_happiness_report = pd.read_csv("worldwide_happiness_report.csv")
world_happiness_report
```

```python
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
world_happiness_report.drop(useless_cols, axis = 1, inplace = True)
```

```python

world_happiness_report.head()
```

```python
world_happiness_report.shape
```

```python
world_happiness_report.set_index("Country or region", inplace = True)
world_happiness_report.head()
```

```python
#joining the datasets first and second
data = corona_data.join(world_happiness_report, how = "inner")
data.head()
```
