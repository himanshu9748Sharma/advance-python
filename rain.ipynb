```python
#importing other necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather_final.csv") 
data
```

```python
X = data.drop('PrecipitationSumInches', axis = 1)
Y = data['PrecipitationSumInches']
```

```python
#reshaping the data
Y = np.asarray(Y).reshape(-1,1)
```

```python
Y
```

```python
day_index = 798
days = [ i for i in range(Y.size)]
```

```python
#intialize the classifier regressor
clf = LinearRegression()
#Training the model
clf.fit(X,Y)
```

```python
#plotting the graph
print("The precipitation prediction graph:")
plt.scatter(days, Y, color = "blue")
plt.scatter(days[day_index], Y[day_index], color = "red") 
plt.title("Precipitation Level")
plt.xlabel("days")
plt.ylabel("Precipitation in inches")
X_vis = X.filter(["TempAvgF", "DewPointAvgF", "WindAvgMPH", "VisibilityAvgMiles", "SeaLevelPressureAvgInches","HumidityAvgPercent"])
plt.show()
```

```python
print("The precipitation vs attribute trend graph:")

for i in range (X_vis.columns.size):
    plt.subplot(2,3,i+1)
    plt.scatter(days, X_vis[X_vis.columns[i][:100]], color = "green")
plt.scatter(days[day_index], X_vis[X_vis.columns.values[i]][day_index], color = "red")
plt.title(X_vis.columns[i])
plt.show()
```
